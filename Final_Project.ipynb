```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
df.info()
```

```python
df.fillna(method='bfill')
```

```python
df.isnull().sum()/df.shape[0]*100
```

```python
df['condition']=df['condition'].fillna(df['condition'].mode())
```

```python
mean=df['price'].mean()
df['price']=df['price'].fillna(mean)
```

```python
std=df['price'].std()
df=df[(df['price']<=mean+3*std)&(df['price']>=mean-3*std)]
```

```python
mean,std
```

```python
df.groupby('region').groups
```

```python
df_agg=df.groupby('region').agg({'price':['mean','min','max']})
```

```python
plt.scatter(df['year'],df['price'])
plt.xlabel('year')
plt.ylabel('price')
plt.title('scatter plot')
plt.legend(['correlation'])
plt.show()
```

```python
df['price'].value_counts().plot(kind='bar',color='blue')
plt.xlabel('price')
plt.ylabel('frequency')
plt.title('bar plot')
plt.legend(['price'])
plt.show()
```
